```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
metrics = pd.read_csv('metrics.csv')
```

```python
metrics
```

```python
%matplotlib inline
```

```python
slope = []
for x in range(0, 30):
    slope.append(0.0199 * x + 0.017)
```

```python
fig, ax1 = plt.subplots()

plt.xlabel('puzzles')
plt.tight_layout()
ax2 = ax1.twinx()
#ax3 = ax1.twinx()
a=ax1.bar(metrics['puzzle'], metrics['accuracy'], color='blue', align='center', width=0.75)
b=ax2.plot(metrics['normalized difficulty'], color="red")
c=ax2.plot(slope, color="orange")

ax1.set_xticklabels(metrics['puzzle'],rotation=45, ha='right', fontsize=9.5)
lns = [a]+b+c
plt.legend(lns,["accuracy","difficulty","difficulty regression curve"], loc=0, bbox_to_anchor=(1.06, 1))
#ax2.set_ylabel('Your second Y-Axis Label goes here!')
fig.savefig('puzzle_acc_diff.pdf', bbox_inches="tight")
```

```python
users = pd.read_csv('users.csv')
```

```python
users
```

```python
N=57
colors = np.random.rand(N)

accuracy = users['accuracy']
active = users['active time avg']
puzzles = users['puzzles number']
competency = users['competency']

plt.rcParams.update({'font.size': 20})

plt.figure(figsize=(18,10), dpi=80)
plt.scatter(accuracy, competency, c=puzzles, cmap='Blues', linewidth=5, alpha=0.8)

plt.clim(12, 30)
cbar = plt.colorbar()
cbar.set_label('puzzles solved')


plt.xlabel('accuracy')
plt.ylabel('competency')

plt.savefig('user_metrics.pdf')
```
